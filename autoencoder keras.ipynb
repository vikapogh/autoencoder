```python
from keras.layers import Input, Dense,Dropout,InputLayer,GaussianNoise
from keras.models import Model,Sequential
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
```

```python
encoding_dim = 32
hidden_layer=100
input_img = Input(shape=(784,))
g=GaussianNoise(0.1)(input_img)
d=Dropout(0.5)(g)
middle1 = Dense(hidden_layer, activation='tanh')(d)
encoded = Dense(encoding_dim, activation='relu')(middle1)
middle2 = Dense(hidden_layer, activation='tanh')(encoded)
decoded = Dense(784, activation='sigmoid')(middle2)
autoencoder = Model(input_img, decoded)
```

```python
encoder = Model(input_img, encoded)
```

```python
autoencoder.compile(optimizer='adam',  loss='binary_crossentropy')
```

```python
(x_train, _), (x_test, _) = mnist.load_data()
```

```python
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))
print x_train.shape
print x_test.shape
```

```python


autoencoder.fit(x_train, x_train,
                epochs=20,
                batch_size=32,
                shuffle=True,
                validation_data=(x_test, x_test))

```

```python

encoded_imgs = encoder.predict(x_test)
decoded_imgs = autoencoder.predict(x_test)
```
